###Basic Python Syntax: Loops

```python
x = 7
while x > 0:
  print("positive x=" + str(x))
  x = x - 1

print("now x=" + str(x))
```

```python
for x in range(3):
  print("x=" + str(x))
```

###Basic Python Syntax: break & continue

Break

```python
for x in range(3):
  print("x=" + str(x))
  if x == 1:
    break
```

Continue

```python
for x in range(3, 0, -1):
  if x % 2 == 0:
    continue
  print(x)
```
